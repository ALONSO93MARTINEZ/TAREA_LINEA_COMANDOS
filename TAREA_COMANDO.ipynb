```python
import time
import random

def selection_sort(arr):
    start = time.time()
    sort(arr)
    end = time.time()
    print("Selection Sort Time: ", round(end - start, 5), "Seconds")

def built_sort(arr):
    start_builting = time.time()
    arr.sort()
    end_builting = time.time()
    print("Built-in Sort Time: ", round(end_builting - start_builting, 5), "Seconds")
```

```python
arr = [random.randint(1, 10000) for _ in range(10000)]
selection_sort(arr)
built_sort(arr)
```

```python
def fibonacci(n):
    x = 0
    y = 1
    fibonacc = [x, y]
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else :
      for i in range(n):
          x, y = y, x + y
          fibonacc.append(y)
      print(x)
    y = 1
```

```python
import time

n = 120
start_time = time.time()
fib_result = fibonacci(n)
end_time = time.time()

print(f"Fibonacci({n}) =", fib_result)
print("Execution Time:", end_time - start_time)
```

```python
import nltk
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

nltk.download('punkt_tab')
nltk.download('stopwords')


texto = "Hablar un idioma es asumir un mundo, una cultura. Nada desarrolla tanto la inteligencia como viajar. La aventura vale la pena"
tokens = nltk.word_tokenize(texto.lower())
stop_words = set(nltk.corpus.stopwords.words('spanish'))
tokens_filtrados = [word for word in tokens if word.isalnum() and word not in stop_words]
frecuencia_palabras = Counter(tokens_filtrados)

plt.figure(figsize=(10, 5))
plt.bar(frecuencia_palabras.keys(), frecuencia_palabras.values())
plt.title('Palabras más frecuentes')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.show()

wordcloud = WordCloud(width=500, height=250, background_color='white').generate_from_frequencies(frecuencia_palabras)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
